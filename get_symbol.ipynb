```python
from google.colab import files
import pandas as pd
import json

f = open('symbols_AMEX.json')
data = json.load(f)
len_symbols = len(data)
print(f"len_symbols_AMEX=", len_symbols)

symbols_AMEX = []
for i in range(len_symbols):
  symbols_AMEX.append(data[i]['symbol'])

f.close()

df = pd.DataFrame(symbols_AMEX)
print(f"df_length=", len(df))

filename = "Symbols_AMEX.csv"
df.to_csv(filename, index=False)

files.download(filename)
```
